```python
# Importing pandas

import pandas as pd
```

```python
# Use a twitter text-classification model from huggingface

from transformers import pipeline

pipe = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-sentiment-latest")
```

```python
# Input data 

df = pd.read_csv('flipkartReviews.csv')
df
```

```python
# Taking only 2000 rows for speeding up the process

df1 = df.iloc[:2000,:]
```

```python
# Creating new column 'Sentiment' and assigning 0 for each record of the column

df1['Sentiment'] = 0
```

```python
# data with new column named 'Sentiment'

df1
```

```python
# Running Model on each review in the data using loop and storing it in the list l

l = []
for reviews in df1['Summary']:
    l.append(pipe(reviews)[0]['label'])
```

```python
# Assigning all the values of l(sentiments) to the column 'Sentiment'

df1['Sentiment'] = l
```

```python
# Presenting data with the final result

df1
```

```python
# Counting numbers for each class(+ve,-ve, neutral) 

df1['Sentiment'].value_counts()
```
